```python
freq = df1['ALIAS_PATIENT'].value_counts().reset_index()
freq.columns = ['ALIAS_PATIENT', 'SICKNESS_COUNT']
print(freq)
```

```python
merged = pd.merge(
    df1, df2,
    how="left", 
    left_on="ALIAS_PATIENT", 
    right_on="Name"
)
```

```python
result = merged[["ALIAS_PATIENT", "DIAGNOSIS DESC", "DUE_TOTAL", 
                 "Department", "Division", "Position", "Location"]]
```

```python
cleaned = result.dropna()
```
